- [X] 弄清楚各个字段都是什么意思
    - SHIPMENT ONBOARD DATE: 实际离开起运港
    - ESTIMATED ONBOARD DATE：预计离开起运港日期
    - DAILY TRACK AND TRACE：日常检测（可省略）
    - TRANSIT PORT ETA：预计到达中转港时间
    - TRANSIT PORT ETD：预计离开中转港时间
    - TRANSIT PORT ATD：实际离开中转港时间
    - PICKED UP：货物被提
    - CARGO ARRIVAL AT DESTINATION：货物到达终点
    - ARRIVAL AT PORT：*实际到达目的港*（非常重要）
    - TRANSIT PORT ATA：实际到达中转港时间
    - PLANNED PICK UP DATE：计划提货时间（可忽略）
    - UPDATE SHIPMENT ETA：更新的预计货运到达时间（可忽略）
    - ESTIMATED ARRIVAL TIME TO PORT：预计到达港口时间（可忽略）
    - ARRIVAL AT CFS OR CY：到达集装箱货运站或集装箱堆场
    - IMP CUSTOMS CLEARANCE START：进口报关开始
    - IMP CUSTOMS CLEARANCE FINISHED：进口报关结束
    - DISCHARGED：卸货时间（可忽略）
    - RDC ATD：区域配送中心实际离开时间（可忽略）
- [X] 保留有效的字段
    - TRANSIT PORT ATD实际离开中转港
    - SHIPMENT ONBOARD DATE实际离开起运港
    - TRANSIT PORT ATA实际到达中转港
    - ARRIVAL AT PORT实际到达目的港
    - (CARGO ARRIVAL AT DESTINATION：货物到达终点)：这个时间通常与ARRIVAL AT PORT很接近
